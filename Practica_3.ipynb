```python
def completa():
    G0 = nx.complete_graph(100)
    nx.draw_networkx(G0)
    plt.title('Red Completa')
    plt.show()
    
    print("La cantidad de nodos en la red es de: ",G0.number_of_nodes())
    print("La cantidad de enlaces en la red es de: ",G0.number_of_edges())
    print("El diametro de la red es: ", nx.diameter(G0))
    print("La densidad de la red es: ",nx.density(G0))
    
    grados = [G0.degree(n) for n in G0.nodes()]
    
    return(grados)
```

Función de la libreria networkx que usa el metodo de erdos donde los datos que nos pide
es el numero de nodos y la probabilidad de enlace entre estos.

```python
def erdos():
    G1 = nx.erdos_renyi_graph(100,0.06)
    plt.figure(figsize = (7,7))
    plt.title('Red Aleatoria')
    nx.draw(G1, with_labels=True)
    plt.show()
    
    print("La cantidad de nodos en la red es de: ",G1.number_of_nodes())
    print("La cantidad de enlaces en la red es de: ",G1.number_of_edges())
    print("El diametro de la red es: ", nx.diameter(G1))
    print("La densidad de la red es: ",nx.density(G1))
    
    grados2 = [G1.degree(n) for n in G1.nodes()]
    
    return(grados2)
```

Función de la libreria networkx que usa el metodo de Watts-Strogatz del mundo pequeño donde solo ponemos
el numero de nodos, grado inicial y probabilidad de enlace.

```python
def mundo():
    G2 = nx.watts_strogatz_graph(n=100, k=4, p=0.3)
    pos = nx.circular_layout(G2)
    plt.figure(figsize = (7,7))
    plt.title('Mundo Pequeño')
    nx.draw_networkx(G2,pos)
    
    print("La cantidad de nodos en la red es de: ",G2.number_of_nodes())
    print("La cantidad de enlaces en la red es de: ",G2.number_of_edges())
    print("El diametro de la red es: ", nx.diameter(G2))
    print("La densidad de la red es: ",nx.density(G2))
    
    grados3 = [G2.degree(n) for n in G2.nodes()]
    
    return(grados3)
```

Función de libreria networkx que usa el metoso de red de escala libre de Barabasi-Albert donde solo se nos pide
el numero de nodos y la cantidad de enlaces por cada nuevo nodo agregado a la red.

```python
def escala():
    G3=nx.barabasi_albert_graph(n=100, m=2)
    plt.figure(figsize = (7,7))
    plt.title('Red de Escala Libre')
    nx.draw_networkx(G3)
    plt.show()
    
    print("La cantidad de nodos en la red es de: ",G3.number_of_nodes())
    print("La cantidad de enlaces en la red es de: ",G3.number_of_edges())
    print("El diametro de la red es: ", nx.diameter(G3))
    print("La densidad de la red es: ",nx.density(G3))
    
    grados4 = [G3.degree(n) for n in G3.nodes()]
    
    return(grados4)
```

Habiendo sacado la distribución de grados de cada red generada lo mandamos a la siguiente función
para que por medio de una grafica de barras grafique las 4 distribuciones en una sola.

```python
def distr_grados(grados,grados2,grados3,grados4):
    plt.figure(figsize = (6,6))
    plt.hist(grados)
    plt.hist(grados2)
    plt.hist(grados3)
    plt.hist(grados4)
    plt.title('Distribución de Grados')
    plt.show()
```

```python
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as plt

grados = completa()
grados2 = erdos()
grados3 = mundo()
grados4 = escala()

distr_grados(grados,grados2,grados3,grados4)
```
