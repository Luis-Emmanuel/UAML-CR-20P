```python
def crea_un_grafo_completo(lst_nodos):
    red = {}

    for i in range(1,n+1):              
        lst_temp = set(lst_nodos) 
        lst_temp.remove(i) 
        up = {i:lst_temp}  
        red.update(up)
    print('La red resultante es:\n',red)
```

```python
def calcula_grados_entrada(d_grafica):
    grado_e = {}

    for i in d_grafica:
        up = {i:len(d_grafica[i])}   
        grado_e.update(up)
    print(grado_e)
    return(grado_e)
```

```python
def dist_grados_entrada(grado_e):
    distr = {}
    lst_temp2 = list()

    for values in grado_e.values():      
        lst_temp2.append(values)
        lst_temp2.sort() 
    
    for i in lst_temp2: 
        llave = lst_temp2.count(i) 
        up2 = {llave:i}
        distr.update(up2)
    print(distr)
    return(distr)
```

```python
def dist_normalizada(distr):
    distr_norm = {}

    for i in distr:
        llave = distr[i]
        valor = i
        up3 = {llave:valor}
        distr_norm.update(up3)
    print(distr_norm)
    return(distr_norm)
```

```python
def grafica_distr(distr):
    x = list()
    y = list()

    for i in distr:
        x.append(distr[i])
        y.append(i)
    plt.plot(x,y)
    plt.title('Distribución de Grados de Entrada')
    plt.xlabel('grado nodal')
    plt.ylabel('numero de nodos')
    plt.show()
```

```python
def grafica_distr_norm(distr_norm):
    x2 = list()
    y2 = list()

    suma = sum(distr_norm.values())  

    for i in distr_norm:
        x2.append(distr_norm[i]) 
        valor = distr_norm[i]/suma 
        y2.append(valor)
    plt.plot(x2,y2)
    plt.title('Distribución Normalizada de Grados de Entrada')
    plt.xlabel('grado nodal')
    plt.ylabel('proporción de nodos')
    plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt


d_grafica = {1:[2], 2:[4], 3:[2], 4:[1,3,5], 5:[2,4,6], 6:[1]}
grado_e = calcula_grados_entrada(d_grafica)
distr = dist_grados_entrada(grado_e)
distr_norm = dist_normalizada(distr)
grafica_distr(distr)
```

```python
grafica_distr_norm(distr_norm)
```
